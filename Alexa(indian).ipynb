```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
from googlesearch import search

# Initialize text-to-speech engine
engine = pyttsx3.init()
engine.setProperty('rate', 150)  # Speed of speech
engine.setProperty('volume', 0.9)  # Volume level

# Set Alexa-like voice
voices = engine.getProperty('voices')
for voice in voices:
    if "english" in voice.name.lower() and "female" in voice.gender.lower():
        engine.setProperty('voice', voice.id)
        break

def speak(text):
    """Function to convert text to speech."""
    engine.say(text)
    engine.runAndWait()

def listen():
    """Function to recognize voice commands."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source, timeout=9, phrase_time_limit=8)
            command = recognizer.recognize_google(audio)
            print(f"You said: {command}")
            return command.lower()
        except sr.UnknownValueError:
            return None
        except sr.RequestError:
            speak("There was an issue connecting to the speech service.")
            return None
        except Exception as e:
            print(f"Error: {e}")
            return None

def search_google(query):
    """Search Google and return the top result."""
    speak("Searching Google...")
    try:
        # Ensure the result is a list, not a generator
        search_results = list(search(query, num_results=1))  # Convert generator to list
        if search_results:
            result = search_results[0]
            print(f"Top result: {result}")
            speak(f"The top result is: {result}")
        else:
            speak("No results found.")
    except Exception as e:
        speak(f"An error occurred while searching: {e}")

def play_youtube(query):
    """Play a YouTube video based on the query."""
    speak(f"Playing {query} on YouTube...")
    pywhatkit.playonyt(query)

def wake_word_detected(command):
    """Check if the wake word 'Alexa' is in the command."""
    return "alexa" in command

def main():
    """Main function for the assistant."""
    speak("Hello, I am Alexa. Say my name to activate me.")
    while True:
        command = listen()
        if command and wake_word_detected(command):
            # Remove "Alexa" from the command
            command = command.replace("alexa", "").strip()
            if command:
                if "play" in command:
                    # Play a YouTube video
                    video_query = command.replace("play", "").strip()
                    play_youtube(video_query)
                elif "search" in command:
                    # Search on Google
                    search_query = command.replace("search", "").strip()
                    search_google(search_query)
                elif "exit" in command or "stop" in command:
                    # Exit the assistant
                    speak("Goodbye! Have a nice day.")
                    break
                else:
                    speak("Sorry, I can't handle that command yet.")
            else:
                speak("What can I do for you?")
        else:
            # If the wake word isn't detected, keep listening
            print("Wake word not detected, continuing...")
            continue

if __name__ == "__main__":
    main()
```
